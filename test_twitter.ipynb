```python
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECLET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)
api = tweepy.API(auth)
#-------------------------------------------------------------------
 
#ツイートを投稿
Account = "54764039" #取得したいユーザーのユーザーIDを代入
tweets = api.user_timeline(Account, count=100, page=1)
num = 1 #ツイート数を計算するための変数
for tweet in tweets:
    print('user : ', tweet.user.screen_name)  # ユーザー名
    print(tweet.text)                  # ツイート内容
    print('='*80) # =を80個表示
    num += 1 # ツイート数を計算
```
